### Git and DVC Install
1. https://git-scm.com/download/win
2. https://dvc.org/#use-cases

DVC is used for large size of data where we can't store it into git. Since git has a space limitation, we push the large-size data into the cloud/another platform like S3 bucket and version them by DVC. Whenever we require, we can pull the data from S3 backet by DVC based on version

DVC keeps metafiles in Git instead of Google Docs to describe and version control your data sets and models. DVC supports a variety of external storage types as a remote cache for large files.

**DVC tracks ML models and data sets**
DVC is built to make ML models shareable and reproducible. It is designed to handle large files, data sets, machine learning models, and metrics as well as code.
